```python
def fetch_agenda_text_from_url(url):
    """Scrapes visible text from the agenda web page."""
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Get only the main content (simplified: all text)
    page_text = soup.get_text(separator="\n")
    
    # Optional: Clean up empty lines
    lines = [line.strip() for line in page_text.splitlines() if line.strip()]
    return "\n".join(lines)

def save_text_to_pdf(text, output_path="Agenda.pdf", title="City Council Meeting Agenda"):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_auto_page_break(auto=True, margin=15)

    pdf.set_font("Arial", "B", 14)
    pdf.multi_cell(0, 10, title)
    pdf.ln()

    pdf.set_font("Arial", size=12)
    for line in text.splitlines():
        pdf.multi_cell(0, 8, line)
        pdf.ln(0.5)

    pdf.output(output_path)
    print(f"Saved agenda to {output_path}")
```

```python
# made it a function (extracting text from PDF)
def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file."""
    try:
        with open(pdf_path, "rb") as file:
            reader = PyPDF2.PdfReader(file)
            text = "\n".join([page.extract_text() or "" for page in reader.pages])
            logger.info(f"Successfully extracted {len(text)} characters from PDF")
            return text
    except FileNotFoundError:
        logger.error(f"PDF file not found: {pdf_path}")
        raise
    except Exception as e:
        logger.error(f"Error reading PDF file: {e}")
        raise

def extract_text_from_url(url):
    """Extract and clean text from a webpage."""
    try:
        logger.info(f"Fetching content from URL: {url}")
        response = requests.get(url, timeout=30)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Remove all script, style, and other non-content elements
        for element in soup(["script", "style", "header", "footer", "nav"]):
            element.extract()
            
        # Focus on the main content area if possible
        main_content = soup.find("div", class_="LegistarContent") or soup
        
        # Get text and clean it
        text = main_content.get_text(separator='\n', strip=True)
        
        # Clean up extra whitespace and normalize
        text = re.sub(r'\n+', '\n', text)
        text = re.sub(r'\s+', ' ', text)
        
        logger.info(f"Successfully extracted {len(text)} characters from URL")
        return text
    except Exception as e:
        logger.error(f"Error fetching or parsing URL content: {e}")
        raise
```

```python
def extract_agenda_from_marker(pdf_path, start_marker="PUBLIC HEARINGS/GENERAL BUSINESS", end_marker="NON-AGENDA ITEMS AND COMMENTS"):
    """
    Extracts agenda-relevant text from a PDF between the start_marker and end_marker.
    """
    doc = fitz.open(pdf_path)
    full_text = ""

    for page in doc:
        full_text += page.get_text()

    upper_text = full_text.upper()
    start_index = upper_text.find(start_marker.upper())
    end_index = upper_text.find(end_marker.upper())

    if start_index == -1:
        print(f"Start marker '{start_marker}' not found in agenda PDF.")
        return full_text.strip()  # fallback to full text

    if end_index == -1:
        print(f"End marker '{end_marker}' not found. Returning everything from start marker onward.")
        trimmed_text = full_text[start_index:]
    else:
        trimmed_text = full_text[start_index:end_index]

    return trimmed_text.strip()

def summarize_agenda_items(pdf_path):
    agenda_text = extract_agenda_from_marker(pdf_path)

    prompt = f"""
You are given an agenda from a city council meeting.

Please summarize it into a **brief, numbered list of distinct agenda topics**. (1., 2., 3.) Each topic should be phrased in a short sentence or phrase (no long paragraphs). There should be one line per topic.

Only list the main items to be discussed or voted on — do not include headers or procedural content.

**Agenda items usually appear in the agenda as numeric labels like '25-0101' or '25-0054'.**

Agenda:
{agenda_text}
"""

    response = model.generate_content(prompt)
    lines = [line.strip() for line in response.text.strip().split("\n") if line.strip()]
    return lines

def save_agenda_summary_to_pdf(agenda_points, output_path="agenda_summary.pdf", meeting_date="City Council Meeting Agenda Summary"):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.set_font("Arial", "B", 14)
    pdf.multi_cell(0, 10, meeting_date)
    pdf.ln()

    pdf.set_font("Arial", size=12)
    for point in agenda_points:
        pdf.multi_cell(0, 8, point)
        pdf.ln(1)

    pdf.output(output_path)
```

```python
def extract_meeting_date(pdf_path):
    """Extracts a date string like 'Tuesday, January 28, 2025' from the first page of a PDF."""
    doc = fitz.open(pdf_path)

    # Grab the first page of the agenda only
    first_page_text = doc[0].get_text()

    # Regex pattern for weekday, month, day, year (e.g., Tuesday, January 28, 2025)
    date_pattern = r"\b(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday),\s+(?:January|February|March|April|May|June|July|August|September|October|November|December)\s+\d{1,2},\s+\d{4}\b"

    match = re.search(date_pattern, first_page_text)
    if match:
        meeting_date = match.group()
        return meeting_date
    else:
        print("No date found in agenda PDF.")
        return None
```

```python
def extract_pdf_text(path):
    """Extracts and concatenates all text from a PDF file."""
    doc = fitz.open(path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

def chunk_text(text, max_tokens=3000):
    """
    Splits text into chunks of approximately max_tokens (words).
    Gemini handles ~8k tokens max per request, so use ~3k-4k for safety.
    """
    sentences = text.split('. ')
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk.split()) + len(sentence.split()) < max_tokens:
            current_chunk += sentence + ". "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + ". "
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks
```
